```python
null_counts = df.isnull().sum()
null_counts
# Customer Lname, Customer Zipcode and Order Zipcode have NULL value
# but can leave it for now
```

```python
# delete column "Product Description", "Product Image", "Benefit per order", "Order Item Total", "Order Item Product Price"
df.drop(columns=['Product Description','Customer Email', 'Customer Password', 'Product Image', 'Benefit per order', 'Order Item Total', 'Order Item Product Price'],  inplace=True)
```

```python
# change shipping date, order date, Days for shipping (real), Days for shipment (scheduled) to date format
df.rename(columns={'shipping date (DateOrders)': 'Shipping Date'}, inplace=True)
df.rename(columns={'order date (DateOrders)': 'Order Date'}, inplace=True)

df['Shipping Date'] = pd.to_datetime(df['Shipping Date'])
df['Order Date'] = pd.to_datetime(df['Order Date'])

# change
column_list = ['Category Id', 'Customer Id', 'Customer Zipcode', 'Department Id', 'Order Customer Id', 'Order Id', 'Order Item Cardprod Id', 'Order Item Id', 'Order Zipcode', 'Product Card Id', 'Product Category Id', 'Product Status']
for cln in column_list:
    df[cln] = df[cln].astype('object')

```
